```python
#next
Employees = pd.DataFrame({'ID': [1, 2, 3, 4, 5],
    'Name': ['Sara', 'Katie', 'David', 'Hazal', 'Georgie'],
    'Address': [42, 38, 36, 41, 35],
    'Salary': [33000, 45000, 65000, 95000, 35000]
})
Employees
```

```python
Department = pd.DataFrame({
    'ID' :[1,3,5,7],
     'Course':['Sales','HR','Marketing','Finance']
})
Department
```

```python
#inner join - all matching rows
inner_join_df=pd.merge(Employees,Department,on="ID",how="inner")
inner_join_df
```

```python
#outerjoin=> all rows from first and second
outer_join_df=pd.merge(Employees, Department, on ="ID", how="outer")
outer_join_df
```

```python
#left join -> all rows from first and matching from second
left_join_df=pd.merge(Employees,Department, on ="ID", how="left")
left_join_df
```

```python
#right join -> all rows from second and matching from first
right_join_df=pd.merge(Employees, Department, on ="ID", how="right")
right_join_df
```

```python
#concatenate
concat_df=pd.concat([Employees,Department],axis=1)
concat_df
```
